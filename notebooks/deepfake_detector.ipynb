## Step 1 — Setup

```python
import os, sys

# Add src/ to path so we can import our modules
sys.path.append('/kaggle/input/deepfake-detector-system/src')

import warnings
warnings.filterwarnings('ignore')

import tensorflow as tf
tf.get_logger().setLevel('ERROR')

print('✅ TensorFlow:', tf.__version__)
print('✅ GPU devices:', len(tf.config.list_physical_devices('GPU')))
```

## Step 2 — Verify Dataset

```python
from data_loader import verify_dataset, build_generators

verify_dataset()
```

## Step 3 — Build & Train Model

```python
from model import build_model
from train import set_seeds, train

set_seeds(42)

train_gen, val_gen, test_gen = build_generators()

model = build_model()
model.summary()
```

```python
history = train(model, train_gen, val_gen)
print('\n✅ Training complete!')
```

## Step 4 — Evaluate on Test Set

```python
from evaluate import evaluate

evaluate(model, test_gen)
```

## Step 5 — Plot Training History

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.patch.set_facecolor('#0b1220')

for ax in axes:
    ax.set_facecolor('#111827')
    ax.tick_params(colors='white')
    ax.xaxis.label.set_color('white')
    ax.yaxis.label.set_color('white')
    ax.title.set_color('white')
    for spine in ax.spines.values():
        spine.set_edgecolor('#243041')

# Accuracy
axes[0].plot(history.history['accuracy'],     color='#10b981', linewidth=2, label='Train')
axes[0].plot(history.history['val_accuracy'], color='#8b5cf6', linewidth=2, label='Validation')
axes[0].set_title('Accuracy',  fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend(facecolor='#111827', labelcolor='white')
axes[0].grid(True, alpha=0.2)

# Loss
axes[1].plot(history.history['loss'],     color='#ef4444', linewidth=2, label='Train')
axes[1].plot(history.history['val_loss'], color='#f97316', linewidth=2, label='Validation')
axes[1].set_title('Loss',  fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend(facecolor='#111827', labelcolor='white')
axes[1].grid(True, alpha=0.2)

plt.suptitle('Training History', fontsize=14, fontweight='bold', color='white')
plt.tight_layout()
plt.show()
```
